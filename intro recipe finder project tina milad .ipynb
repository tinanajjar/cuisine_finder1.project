```python
### App logo designed from https://www.asciiart.eu/text-to-ascii-art################
###################################################################################################
logobook = r"""
██████╗ ███████╗██╗   ██╗████████╗
██╔══██╗██╔════╝██║   ██║╚══██╔══╝
██████╔╝███████╗██║   ██║   ██║   
██╔═══╝ ╚════██║██║   ██║   ██║   
██║     ███████║╚██████╔╝   ██║   
╚═╝     ╚══════╝ ╚═════╝    ╚═╝   
"""
############################################################
################ Import file as Read 'r' ###################
with open('recipe_data.json', 'r') as f:
    recipe_data = json.load(f)
    


# Create list of available recipe from dict
def recipe_list():
    list = []
    for x in recipe_data:
        list.append(x)
    return list


# Create List with available cuisine from dict
def cuisine():
    cuisine_list = []
    for recipe in recipe_data:
        cuisine_list.append(recipe_data[recipe]['Cuisine'])
    return list(set(cuisine_list))  # Use Set to Remove Duplicate name from cuisine list


# Create List with available Prep Time from dict
def preptime():
    PrepTime_list = []
    for recipy in recipe_data:
        x = recipe_data[recipy]['Prep Time']
        if isinstance(x, (int,float)) and x >=0: #Exclude None,String,- elements from preptime list
            PrepTime_list.append(recipe_data[recipy]['Prep Time'])
    return sorted(set(PrepTime_list))  # Use Set to Remove Duplicate from Prep Time list


# Create List with available Prep Time from dict
def rating():
    rating_list = []
    for recipy in recipe_data:
        if recipe_data[recipy]['Rating'] != None:
            rating_list.append(recipe_data[recipy]['Rating'])
    return sorted(set(rating_list))  # Use Set to Remove Duplicate from Prep Time list



#Function to get list of available rate in DB {min/equal,max} from user_input rate. x = 4.5
def close_search(x,y): #x for user input rate, y for rate list y = [3.7, 4.0, 4.2, 4.3, 4.4, 4.5, 4.6, 4.7, 4.8, 4.9]
    below = max([i for i in y if i < x], default=None)
    equal = x if x in y else None
    above = min([i for i in y if i > x], default=None)
    result = [i for i in [below, equal, above] if i is not None] 
    return result


#Creat function to get recipe for specific cuisine:
def recipe_cuisine(cuisine):
    # using .items() give both (name like 'Chicken Kabsa' and details like Dictionary of its info 
                     #(e.g., {'Cuisine': 'Italian', 'Ingredients': [...], ...}) in each loop
    
    matches = [name for name, details in recipe_data.items() if details['Cuisine'].lower() == cuisine.lower()]
    for i in matches:
        print(f"Try {cuisine} '{i}' cuisine")


# Creat function to get recipe details:
def recipe_details(x):
    match = [recipe_data[x] for i in recipe_data if i == x]
    print(f'''Recipe: {x}
        Cuisine: {recipe_data[x]['Cuisine']}
        Ingredients: {recipe_data[x]['Ingredients']}
        Prep Time: {recipe_data[x]['Prep Time']}
        Difficulty: {recipe_data[x]['Difficulty']}
        Rating: {recipe_data[x]['Rating']}
        ''')



# Create function to filter by x for: ['Difficulty', 'Prep Time', 'Rating'] and 
        #By y : where y refer to string ['Hard', 'Medium', 'Easy']/ Time as integer in minutes/ float rating 
def filter(x, y):
    for i in recipe_data:
        try:
            if recipe_data[i][x] == y:
                print(f'''Cuisine: {recipe_data[i]['Cuisine']}
        Ingredients: {recipe_data[i]['Ingredients']}
        Prep Time: {recipe_data[i]['Prep Time']}
        Difficulty: {recipe_data[i]['Difficulty']}
        Rating: {recipe_data[i]['Rating']}''')
        except NameError as e:
            print(f"ERROR: {e}")


# Create Option list to show available option for user to choose
option_list = {1: 'Get recipe recommendations', 2: 'Search for a recipe', 3: 'Filter recipes (Difficulty/Prep Time/Rating)', 4: 'Add a new recipe', 5: 'Exit'}

answer = True
while answer:
    print(logobook)
    print('Welcome to the Recipe Finder!\n')
    for options in option_list:  # Start loop to Ask User to select from option_list
        print(f"{options}. {option_list[options]}")
    try:                #Use try to catch if user input not int(1-5) and move for exception in line-334
        option = int(input("\nPlease Enter your prefer option from list: ").strip())

        ######## Get recipe recommendations  #######
        if option == 1:  
            answer = 'y'
            while answer == 'y':  #####Loop to continue searches

                user_cuisine = input('Enter your preferred cuisine(Italian/Indian/Arabic/....):  ').strip().title()  # User select cuisine {even partial word} with make upper for first letter
                while not user_cuisine:  ##### Not Allowing Empty field ######
                    print("Cuisine Name Cannot be Empty!!!!")
                    user_cuisine = input('Enter your preferred cuisine(Italian/Indian/Arabic/....):  ').strip().title()
                search_cuisine = [word for word in cuisine() if user_cuisine in word]  # Use partial matching to cuisine
                if search_cuisine: ###if for cuisine exist
                    for word in search_cuisine:
                        print(f"For {word} Cuisine please try following:   ")
                        recipe_cuisine(word)  # Print list of Recipes for specific cuisine inserted by user as input
                    answer = input('Do you want to continue search Y/N: ').strip().lower()  # get reply Y to continue or N to exit
                else:
                    print(f"Sorry No matches found for '{user_cuisine}'")
                    for i in recipe_list():
                        print(f"Based on your love for '{user_cuisine}' Please try '{i}' recipe")  # Print All available recipe in dict
                    answer = input('Do you want to continue search Y/N: ').strip().lower()  # get reply Y to continue or N to exit
            answer ='yes'
            print(f"Thanks you for using Recipe Finder App.! Back to Main List")
        ########################################################################
        ######################## Find a Recipe #################################
        elif option == 2: 
            answer = 'y'
            while answer == 'y':  #####Loop to continue searches

                user_recipe = input('Enter your preferred recipe like (Spagheiti/Shawarma/Pizza/Chiken/...):  ').strip().title()  # User Enter recipe {even partial word} with make upper for first letter
                while not user_recipe:  # Not Allowing Empty field
                    print("Recipe Cannot be Empty!!!!  ")
                    user_recipe = input('Enter your preferred recipe like (Spagheiti/Shawarma/Pizza/Chiken/...):  ').strip().title()  # User Enter recipe {even partial word} with make upper for first letter

                search_recipe = [word for word in recipe_list() if user_recipe in word]  # Use partial matching to recipe
                if search_recipe:
                    for word in search_recipe:
                        #    print(f"For {word} Recipe")
                        recipe_details(word)  # Print Recipes for specific recipe inserted by user as input
                    answer = input('Do you want to continue search Y/N: ').strip().lower()  # get reply Y to continue or N to exit


                else:
                    print(f"Sorry No matches found for '{user_recipe}")
                    for i in recipe_list():
                        print(f"Please try '{i}' recipe")  # Print All available recipe in dict
                    answer = input('Do you want to continue search Y/N: ').strip().lower()  # get reply Y to continue or N to exit

            answer ='yes'  
            print(f"Thanks you for using Recipe Finder App.! Back to Main List")
        ###############################################################################
        ################ Filter recipes (by time/difficulty/rating)####################
        elif option == 3:
            answer = 'y'
            while answer == 'y':

                try:
                    filter_list = ['Difficulty', 'Prep Time', 'Rating']
                    user_filter = input(f'Please Select Filter By ({filter_list}): ').strip().title()  # Remove empty spaces not allowed( User Enter Filter for recipes by (Difficulty/ Prep Time / Rating))
                    while not user_filter:  # Not Allowing Empty field
                        print("Filter cannot be Empty!!!! ")
                        user_filter = input(f'Please Select Filter By ({filter_list}): ').strip().title()  # User Enter Filter for recipes by (Difficulty/ Prep Time / Rating)

                    search_filter = [word for word in filter_list if user_filter in word]
                    if search_filter:
                        for word in search_filter:
                            if word == 'Difficulty': #### Filter with Dificulty####
                                print(f"selected filter '{word}', user entry: '{user_filter}'")
                                difficulty_list = ['Hard', 'Medium', 'Easy']
                                difficulty_level_user = input(f"Please Insert {difficulty_list}: ").strip().title()
                                while not difficulty_level_user:  # Not Allowing Empty field
                                    print("Please Insert difficulty level cannot be empty")
                                    difficulty_level_user = input(f"Please Insert {difficulty_list}: ").strip().title()
                                difficult_filter = [level for level in difficulty_list if difficulty_level_user in level]
                                
                                if difficult_filter:
                                    for level in difficult_filter:
                                        print(word, level)
                                        filter(word, level)
                                        answer = input('Do you want to continue search Y/N: ').strip().lower()  # get reply Y to continue or N to exit
                                else:
                                  print(f"Please Insert difficulty level {difficulty_list} nothing else")  
                                    
                            elif word == 'Prep Time': #### Filter with PrepTime ####
                                print(f"selected filter '{word}', user entry: '{user_filter}'")
                                # let filter recipe by choosing the closest time available in Prep Time list

                                while True:  # Use while to convince user to insert proper time as input
                                    time_user = input(f"Please select Prep Time from list '{preptime()}' in minutes: ").strip()  # to remove space

                                    if not time_user.isdigit():  # only for positive integer entry to avoid (empty,str,float and minus)
                                        print("Please insert a valid number!! integer Only")
                                        continue
                                    time_user = int(time_user)  # convert to integer
                                    break  # Exit loop
                                    
                                closest_time = min(preptime(), key=lambda x: abs(x - time_user))  # Function to calculate abs(x-time_user) how far each number from what user insert and min(...,key=  ) pick the smallest distance
                                print(f"The Closest available Prep Time to {time_user} is {closest_time}")
                                for time in preptime():  # This function to get all recipe which Prep Time is <= to closest_time for user input
                                    if time == closest_time:
                                        filter(word, time)
                                answer = input('Do you want to continue search Y/N: ').strip().lower()  # get reply Y to continue or N to exit


                            elif word == 'Rating': #### Filter with Rating ####
                                print(f"selected filter '{word}', user entry: '{user_filter}'")
                                while True:
                                    rate_user = input(f"Please Insert Rate from list {rating()}:  ")

                                    try:
                                        rate_user = float(rate_user) # to avoid empty,str 
                                        if rate_user <= 0:  # Not allowed 0 or - number
                                            print("Please insert rate from list (Zero/Minuse) Number not allowed")
                                            continue
                                        break  # Exit when positive float valid

                                    except ValueError:
                                        print("Please insert a valid input")
                                        continue
                                        
                                try:
                                    for rate in close_search(rate_user,rating()): # Read rate from list
                                        filter(word, rate)  # Call filter Function with identifier = Rating and rate from list
                                except ValueError:
                                     print(f"Please insert higher than {rate_user}")
                                answer = input('Do you want to continue search Y/N: ').strip().lower()  # get reply Y to continue or N to exit

                            else:
                                print(f"Input '{user_filter}' not Accepted, Select correct input from list ('Difficulty', 'Prep Time', 'Rating') ")
                                answer = input('Do you want to continue search Y/N: ').strip().lower()  # get reply Y to continue or N to exit
                    else:
                        print(f"Input '{user_filter}' not Accepted, Select correct input from list ('Difficulty', 'Prep Time', 'Rating') ")
                        answer = 'n'
                except ValueError:
                    print(f"Please Enter correct Value from list ('Difficulty', 'Prep Time', 'Rating')")
            print(f"Thanks you for using Recipe Finder App.! Back to Main List")
        ##########################################################################
        ############### Add a new Recipe to DataBase:#############################
        elif option == 4:
            answer = 'y'
            while answer == 'y':
                try:
                    ##Insert Recipe Name as mandatory field
                    a = input("Please Insert Recipe Name Example:'Shawrma':   ").title()
                    while not a:
                        print("Please Insert Recipe !!!! Cannot be empty")
                        a = input("Please Insert Recipe Name Example:'Shawrma':   ").title()

                    ##Insert Cuisine as mandatory field
                    x = input("Please Insert Cuisine Type like 'Arabic': ").title()
                    while not x:
                        print("Please Insert Cuisine Type !!!! Cannot be empty")
                        x = input("Please Insert Cuisine Type like 'Arabic': ").title()

                    ##Insert Ingredients as mandatory field max five ingredients and no empty input
                    temp_y = []
                    for num in range(5):
                        items = input("Please Insert Ingredients Type like 'Pasta', 'Eggs', 'Cheese', 'Bacon': ").strip().title()
                        while not items:
                            print("Please Insert!!!! Ingredient cannot be empty")
                            items = input("Please Insert Ingredients Type like 'Pasta', 'Eggs', 'Cheese', 'Bacon': ").strip().title()
                        temp_y.append(items)
                        num += 1
                    y = temp_y

                    #######################################################################
                    ##Insert PrepTime if missing/invalid insert as default value 20   #####
                    try:
                        z_temp = int(input("Please Insert Prep Time Type Or default value is 20: "))
                        if z_temp == '':
                            z = 20
                        else:
                            z = int(z_temp)
                    except ValueError:
                        z = 20 #In case of any ValueError (string/non-integer) insert as 20
                        
                    ###########################################################################
                    ###Insert Difficulty level like Easy/Medium/Hard default value Medium #####
                    try:
                        w_tmp = input("Please Insert Difficulty level like Easy/Medium/Hard: ").upper()
                        if w_tmp == '':
                            w = 'Medium'
                        elif w_tmp not in ['EASY','MEDIUM','HARD']:
                            w = 'Medium'
                        else:
                            w = w_tmp
                    except ValueError:
                        w = 'Medium' ###In case of any ValueError (Not string) insert as 'Medium'
                    
                    ############################################################################
                    ##Insert Rating level if missing/invalid insert as default value None ######
                    try:
                        v_temp = float(input("Please Insert Rating level like 1-5: "))
                        if v_temp == '' or v_temp <= 0: 
                            v = None
                        else:
                            v = float(v_temp)
                    except ValueError:
                        v = None #In case of any ValueError (string/non-float) insert as None
                except ValueError as e:
                    print(f"Error: {e}")
                ################################################################################################
                ################# #############Save new Entry Recipe as Nist dictionay #########################        
                recipe_data[a] = {'Cuisine': x, 'Ingredients': y, 'Prep Time': z, 'Difficulty': w, 'Rating': v}
                #################################################################################################
                ######################### Update JSON file to store new recipe: #################################
                with open("recipe_data.json", "w") as f:
                    json.dump(recipe_data,f, indent=2)
                        
                answer = input('Do you want to continue Insert new Recipe Y/N: ').strip().lower() #### get reply Y to continue or N to exit
            answer ='yes'
        elif option == 5:
            
            answer = False
        else:
            print(f"Sorry '{option}' Not Accepted! Next Time Please Select Number from 1-5: ")
            answer = False
    except ValueError:
        print(f"Sorry input Not Accepted!! Back to Main List Select Number from 1-5: ")
# #####################################################################
# ########### Message When exite Application ###################

print(f"Thanks you for using Recipe Finder App.! Get Back Again, GoodBye!!!")
```
